```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

```python
df_train = pd.read_csv('../input/tabular-playground-series-sep-2021/train.csv')
```

```python
df_train['KFold'] = -1
```

```python
df_train.head()
```

```python
kf = model_selection.KFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
for fold, (train_indicies, valid_indicies) in enumerate (kf.split(df_train)):
    print(fold, len(train_indicies), len(valid_indicies))
    df_train.loc[valid_indicies, "KFold"] = fold
```

```python
df_train.loc[train_indicies, ["KFold"]]
```

```python
df_train.to_csv('train_folds.csv', index = False)
```
